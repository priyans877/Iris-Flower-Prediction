<h1>Iris Flower Prediction</h1>

<h3><i>Iris flower classification is a very popular machine learning project. The iris dataset contains three classes of flowers, Versicolor, Setosa, Virginica, and each class contains 4 features, ‘Sepal length’, ‘Sepal width’, ‘Petal length’, ‘Petal width’. The aim of the iris flower classification is to predict flowers based on their specific features.</i></h3>


<h3>The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). These measures were used to create a linear discriminant model to classify the species. The dataset is often used in data mining, classification and clustering examples and to test algorithms.</h3>

<h2>Convrting Target feature Object values into numerical values</h2>

```python
#changing name of flowers to numerical value for better performence , without any encoding method
def change(y):
    if y=="Iris-setosa":
        return 1
    elif y=="Iris-versicolor":
        return 2
    else:
        return 3

data["species"]=data["species"].apply(change)
```

```python
data['species'].unique()
```

<h2>Finding Corelation between all Features</h2>

```python
#plotting coorelation of features on species

corela=data.corr(method='pearson')
corela
```

<h3>It is founded that species have high +ve corelation on "sepal_length" ,"petal_length" & "petal_width"<h3>

```python
sns.heatmap(corela,annot=True)
```

<h2>Extracting Target features and other features</h2>

```python
#spliting Targget feature from dataset
y=data["species"]

x=data.drop('species',axis=1)
```

```python
x
```

```python
y
```

<h2>Performing Train Test Split on data</h2>

```python
#performing train_test_split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=7)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

<h2>Using DecisionTreeClassifier model</h2>

```python
#using logistic regression because optput is  numerical
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier = classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred=pd.DataFrame(y_pred)
```

<h2>Using LogisticRegression model</h2>

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()

lr_model.fit(x_train,y_train)
```

```python
y_pred_lr=lr_model.predict(x_test)
```

```python
y_pred_lr=pd.DataFrame(y_pred_lr)
```

<h2>Using RandomForestClassifier model</h2>

```python
from sklearn.ensemble import RandomForestClassifier  
rfc_model= RandomForestClassifier(n_estimators= 10, criterion="entropy")  
rfc_model.fit(x_train, y_train)  
```

```python
y_pred_rfc=rfc_model.predict(x_test)
```

```python
y_pred_rfc=pd.DataFrame(y_pred_rfc)
```

<h2>Using KNeighborsClassifier model</h2>

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier()
knn_model.fit(x_train,y_train)
```

```python
y_pred_knn=knn_model.predict(x_test)
```

```python
y_pred_knn=pd.DataFrame(y_pred_knn)
```

<h2>Evaluating Model accuracies</h2>

```python
#evaluation of of all models

from sklearn.metrics import accuracy_score

#Evaluation of decisiontreeclassifier
print("Accuracy score of DecisionTreeClassifier :", accuracy_score(y_pred,y_test))

print("-"*60)
print("-"*60)

#Evaluation of LogisticRegression
print("Accuracy score of LogisticRegression :", accuracy_score(y_pred_lr,y_test))

print("-"*60)
print("-"*60)

#Evaluation of RandomForestClassifier
print("Accuracy score of RandomForestClassifier :", accuracy_score(y_pred_rfc,y_test))

print("-"*60)
print("-"*60)

#Evaluation of KNeighborsClassifier
print("Accuracy score of KNeighborsClassifier :", accuracy_score(y_pred_knn,y_test))
```

<h2>Hence the DecisionTreeClassifier , LogisticRegression & KNeighborsClassifier have accuracy of 89% which is the best accuracy then RandomForestClassifier Algortihm</h2>
